```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

# scatterplot

```python
tv_hours = [3,5,2,0,5,3,1,4,3,4]  # tv izleme süresi
GPA = [2.7,2.1,3.3,3.4,2.0,3.0,3.6,2.8,3.5,2.6]# not ortalaması
```

```python
plt.scatter(x=tv_hours,y=GPA);  # x ekseninde TV izleme saati (integer), y ekseninde not ortalaması (float)
```

```python
plt.figure(figsize=(8,6))
ax = plt.axes()
ax.scatter(tv_hours, GPA)
sns.regplot(tv_hours, GPA)
ax.set_xlabel("TV Hours")
ax.set_ylabel("GPA")
ax.axis("tight")
plt.show();
```

```python
x=np.array(tv_hours)  # listeler np array e çevirdik
y=np.array(GPA)
```

```python
len(x)*sum(x*y)-sum(x)*sum(y)
```

```python
np.sqrt((len(x)*sum(x**2)-sum(x)**2)*(len(y)*sum(y**2)-sum(y)**2))
```

```python
-71/82.84926           # pearson korelasyon katsayısı-------- negatif ve strong----------
```

```python
stats.pearsonr(x,y)               # % 95 kullanırız    pearsonr
```

```python
r,p=stats.pearsonr(x,y)  #  değer verir ilk kımı "r"  ikincisi "p" değeri
```

```python
r
```

```python
p             #ikincisi "p" değeri  korelasyon anlamlı (popülasyonda)
```

# The Spearman rank-order correlation coefficient is a nonparametric measure of the monotonicity of the relationship between two datasets.
Unlike the Pearson correlation, the Spearman correlation does not assume that both datasets are normally distributed.

```python
stats.spearmanr(x,y)    # % 5 kullanırız
```

```python
stats.kendalltau(x,y)  # bu da var
```

# numpy

```python
np.cov(x,y)           # çok anlamlı değilde negatif yönlü ilişki var
# ilk değer sample variance
```

```python
stats.tvar(GPA) # yukarıdaki sağda alttaki sample varyans 
```

```python
np.var(GPA)    # popülasyon varyans
```

```python
stats.tvar(tv_hours)   # sample varyans 
```

```python
np.corrcoef(tv_hours, GPA)  # Corrcoef de hesaplanan şey bize default olarak Pearson koralasyonu verir.
```
