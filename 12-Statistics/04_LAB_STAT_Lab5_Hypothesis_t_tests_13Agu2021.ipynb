```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Brain size and weight and IQ data (Willerman et al. 1991)

df = pd.read_csv("brain_size.csv", sep = ";", na_values = ".", index_col = 0)
```

```python
df.head()
```

```python
# H0: mu0 = 109

# H1: mu0 > 109
```

```python
df["VIQ"].mean()

# VIQ ortalamas 109'dan byk ama bu iddiay test yapmadan (1sample t test) ortaya atamyorum. 
```

```python
df["VIQ"].std()
```

```python
# t_test = (xbar-mu) / (sigma/kk (n)) 

t_test = (df["VIQ"].mean() - 109) / (df["VIQ"].std() / np.sqrt(df.shape[0]))
```

```python
t_test
```

```python
 #pvalue
1-stats.t.cdf(0.897152958632355,39) 
```

```python
oneSamp=stats.ttest_1samp(df['VIQ'], 109, alternative='greater')
oneSamp
# default olarak alternative = two sided. bu yzden biz alternative  = greater yapyoruz.
```

```python
alpha = 0.05
if oneSamp.pvalue < alpha:
    print("Reject the Null")
else:
    print("Fail to reject")
```

```python
df = pd.read_csv("arsenic.csv")
```

```python
df.head()
```

```python
df["x1"].mean()
```

```python
df["x2"].mean()
```

```python
# H0 : mu1 = mu2
# H1 : mu1 != mu2
```

```python
indTest = stats.ttest_ind(df["x1"], df["x2"])
```

```python
indTest.statistic
```

```python
indTest.pvalue
```

```python
alpha = 0.05

if indTest.pvalue < alpha:
    print("Reject the Null")
else:
    print("Fail to reject")
```
